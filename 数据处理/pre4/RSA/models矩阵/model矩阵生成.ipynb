```python
# 指定要读取的 .mat 文件路径
file_path = "./alltypeRDM.mat"  # 将 "path/to/your/file.mat" 替换为实际的文件路径

# 使用 loadmat 函数读取 .mat 文件
data3 = loadmat(file_path)['rdm']

print(data3)
```

```python
import numpy as np

# 创建一个 16x16 的 ndarray
arr = data3

# 生成循环的行名和列名
row_names = np.tile(np.arange(1, 5), 4)
col_names = np.tile(np.arange(1, 5), 4)

# 在数组的左侧和顶部添加一列和一行
arr_with_names = np.pad(arr, ((1, 0), (1, 0)), mode='constant', constant_values=0)

# 在第一行和第一列分别设置行名和列名
arr_with_names[0, 1:] = col_names
arr_with_names[1:, 0] = row_names

# 打印带有循环行名和列名的数组
print(arr_with_names)
```

```python
# 将行名和列名相同的单元格的值改为0
for i in range(1, arr_with_names.shape[0]):
    for j in range(1, arr_with_names.shape[1]):
        if arr_with_names[i, 0] == arr_with_names[0, j]:
            arr_with_names[i, j] = 0
```

```python
import numpy as np

# 创建一个 16x16 的 ndarray
arr = np.zeros((16, 16))

# 生成循环的行名和列名
row_names = np.tile(np.arange(1, 5), 4)
col_names = np.tile(np.arange(1, 5), 4)

# 在数组的左侧和顶部添加一列和一行
arr_with_names = np.pad(arr, ((1, 0), (1, 0)), mode='constant', constant_values=0)

# 在第一行和第一列分别设置行名和列名
arr_with_names[0, 1:] = col_names
arr_with_names[1:, 0] = row_names


# 将行名和列名相同的单元格的值改为0，其他单元格的值改为1
for i in range(1, arr_with_names.shape[0]):
    for j in range(1, arr_with_names.shape[1]):
        if arr_with_names[i, 0] == arr_with_names[0, j]:
            arr_with_names[i, j] = 0
        else:
            arr_with_names[i, j] = 1

# 打印结果
print(arr_with_names)
```
