```python
from thinkdsp import CosSignal
```

```python
samplerate = 44100
freq_message = 4

carrier_sig = CosSignal(freq=1000, amp=1, offset=0)
carrier_wave = carrier_sig.make_wave(duration=1.5, framerate=samplerate)
carrier_wave.plot()

message_sig = CosSignal(freq=freq_message,amp=1,offset=0)
message_wave = message_sig.make_wave(duration=1.5, framerate=samplerate)
message_wave.plot()
```

```python
modulated = message_wave*carrier_wave
modulated.plot()

modulated.write(filename=str(freq_message)+'_1000Hz.wav')
```

```python
'''
用白噪音 - 正态分布随机抽样

思路：简单地把原来 carrier的数值给替换掉就行。


'''
import random


samplerate = 44100
freq_message = 4


carrier_sig = CosSignal(freq=1000, amp=1, offset=0)
carrier_wave = carrier_sig.make_wave(duration=1.5, framerate=samplerate)
carrier_wave.plot()

# 标准正泰分布
exp = 0
var = 1
for i in range(len(carrier_wave.ys)):
    x = random.normalvariate(exp, var)
    carrier_wave.ys[i] = x
    # print(carrier_wave.ys[i])

message_sig = CosSignal(freq=freq_message,amp=1,offset=0)
message_wave = message_sig.make_wave(duration=1.5, framerate=samplerate)
message_wave.plot()
```

```python
modulated = message_wave*carrier_wave
modulated.plot()

modulated.write(filename=str(freq_message)+'_whiteNoise.wav')
```
